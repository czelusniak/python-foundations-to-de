# Exercise 53 ‚Äî `Series` vs `DataFrame`

## Concept

Two core pandas types:

- **`Series`** ‚Äî a labeled 1D array. Think of it as a single column plus an index.
- **`DataFrame`** ‚Äî a labeled 2D table. Think of it as a dict of Series (columns), sharing the same row index.

```python
import pandas as pd

# Series from a list (default integer index 0..n-1)
s = pd.Series([10, 20, 30])
# 0    10
# 1    20
# 2    30

# Series with a custom index
s = pd.Series([10, 20, 30], index=["a", "b", "c"])
s["b"]            # 20  ‚Äî label-based lookup

# DataFrame from a dict of columns
df = pd.DataFrame({
    "id":   [1, 2, 3],
    "name": ["Alice", "Bob", "Carol"],
})

# DataFrame from a list of dicts (typical when coming from JSON / API)
df = pd.DataFrame([
    {"id": 1, "name": "Alice"},
    {"id": 2, "name": "Bob"},
])
```

### Key vocabulary

- `df.columns` ‚Äî Index of column names
- `df.index` ‚Äî Index of row labels (often `RangeIndex(0..n-1)` unless you set one)
- `df.shape` ‚Äî `(rows, cols)`
- `df["col"]` ‚Äî selects one column ‚Üí returns a Series
- `df[["a", "b"]]` ‚Äî selects multiple columns ‚Üí returns a DataFrame

Understanding "is this a Series or a DataFrame?" is the single most important debugging skill ‚Äî error messages get much easier once you can tell at a glance.
