# Exercise 54 — Inspecting a DataFrame

## Concept

The first thing you do with an unfamiliar DataFrame is **look at it**. Memorize this kit:

```python
df.head(5)           # first 5 rows
df.tail(5)           # last 5 rows
df.sample(5)         # 5 random rows — great for spot-checking

df.shape             # (rows, cols)
df.columns           # column names
df.dtypes            # column → dtype
df.index             # row index

df.info()            # dtypes + non-null counts + memory usage
df.describe()        # summary stats for numeric columns
df.describe(include="object")   # for string/categorical columns

df.isna().sum()      # null count per column — your first DQ check
df.nunique()         # distinct values per column
df["col"].value_counts()        # frequency table for one column
```

In DE, **inspection is the work**. Most pipeline bugs trace back to assumptions about the data that were never verified — types, nullability, cardinality, distributions. Build the muscle of inspecting first.
