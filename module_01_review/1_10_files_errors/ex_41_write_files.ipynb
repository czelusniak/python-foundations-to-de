# Exercise 41 — Writing text files

## Concept

Open in write mode and use `f.write(...)`. `.write()` does **not** append a newline — you do that yourself.

```python
with open("out.txt", "w", encoding="utf-8") as f:
    f.write("hello\n")
    f.write("world\n")
```

Modes:
- `"w"` — truncate and write. **Overwrites the file** if it exists.
- `"a"` — append (creates if missing).
- `"x"` — create exclusively. Raises `FileExistsError` if the file is already there. Safer than `"w"` when you want to prevent accidental overwrites.

Writing many lines: either build a single string with `"\n".join(lines)`, or use `f.writelines(lines_with_newlines)`. `.writelines` does NOT add separators between items either — bring your own.
