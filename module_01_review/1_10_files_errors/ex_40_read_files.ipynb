# Exercise 40 — Reading text files

## Concept

Use `open()` inside a `with` block. The `with` ensures the file is closed even if an exception is raised — never use bare `open()` in real code.

```python
with open("data.txt", "r", encoding="utf-8") as f:
    content = f.read()            # entire file as one string
```

More commonly, you iterate line by line — this streams the file and uses constant memory:

```python
with open("data.txt", encoding="utf-8") as f:
    for line in f:
        line = line.rstrip("\n")
        ...
```

Other helpful methods:
- `f.readlines()` — read all lines into a list (eager; avoid for large files)
- `f.readline()` — read one line

**Always** specify `encoding="utf-8"`. Without it Python uses the platform default — `cp1252` on Windows — which silently mangles non-ASCII text.

The `mode` argument:
- `"r"` — read (default)
- `"rb"` — read binary (for non-text data; returns `bytes`)
