# Exercise 23 — `while`, `break`, `continue`

## Concept

`while` loops repeat as long as a condition is true. Use them when you **don't know in advance** how many iterations you need — e.g., reading until end-of-stream, retrying until success, processing a queue that other code is feeding.

```python
n = 0
while n < 5:
    print(n)
    n += 1                 # don't forget this — infinite loop otherwise
```

Two control statements inside any loop (`for` or `while`):

- **`break`** — exit the loop immediately
- **`continue`** — skip the rest of this iteration, go straight to the next

```python
for x in values:
    if x is None:
        continue          # skip this value
    if x < 0:
        break             # stop the whole loop
    process(x)
```

Rule of thumb: if you know the iterable, prefer `for`. Reach for `while` only when iteration count depends on runtime conditions.
