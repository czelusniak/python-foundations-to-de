# Exercise 22 — `zip()`

## Concept

`zip()` pairs up elements from two or more iterables, position by position:

```python
names  = ["Alice", "Bob", "Carol"]
scores = [88, 63, 76]

for name, score in zip(names, scores):
    print(name, score)
# Alice 88
# Bob 63
# Carol 76
```

If the iterables have different lengths, `zip` **stops at the shortest** — silently. That's a footgun: if you have a 1000-row column and a 999-row column, you'll lose the last row without warning. Use `strict=True` (Python 3.10+) to raise `ValueError` instead:

```python
zip(names, scores, strict=True)
```

Classic uses:
- Pairing parallel columns into rows
- Building a dict from two lists: `dict(zip(keys, values))`
- Iterating over consecutive pairs: `zip(lst, lst[1:])`
