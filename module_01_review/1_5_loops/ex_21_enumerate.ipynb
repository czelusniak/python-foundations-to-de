# Exercise 21 — `enumerate()`

## Concept

When you need both the **index** and the **value** while iterating, `enumerate` is the idiomatic tool. Forget `range(len(lst))` — `enumerate` is shorter, faster to read, and harder to get wrong.

```python
fruits = ["apple", "banana", "cherry"]

for i, fruit in enumerate(fruits):
    print(i, fruit)
# 0 apple
# 1 banana
# 2 cherry

for i, fruit in enumerate(fruits, start=1):   # custom starting index
    print(i, fruit)
# 1 apple
# 2 banana
# 3 cherry
```

In DE you'll use `enumerate` constantly: line numbers when parsing files, row indexes when reporting validation errors, position in a batch, etc.
