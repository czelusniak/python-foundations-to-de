# Exercise 19 — `for` loop over a list

## Concept

A `for` loop walks through every element of an iterable (list, string, tuple, dict, etc.) one at a time. You don't manage an index — Python does it for you.

```python
prices = [10.0, 25.5, 7.25]

for p in prices:
    print(p)
```

The variable `p` is just a name that gets rebound to each element in turn. Pick a name that describes **one** item (`price`, not `prices`).

Loops are how you process every row of a dataset, every line of a file, every record from an API — daily bread in DE.
