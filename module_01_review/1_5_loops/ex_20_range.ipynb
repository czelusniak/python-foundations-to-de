# Exercise 20 — `range()` and indexed iteration

## Concept

`range()` produces a sequence of integers. It's lazy (doesn't build the list in memory), and accepts up to three arguments:

```python
range(5)          # 0, 1, 2, 3, 4
range(2, 7)       # 2, 3, 4, 5, 6        — start, stop
range(0, 10, 2)   # 0, 2, 4, 6, 8        — start, stop, step
range(5, 0, -1)   # 5, 4, 3, 2, 1        — negative step
```

The `stop` value is **exclusive** (Python is consistent: same rule as slicing).

When do you actually need `range`?
- Repeating something N times: `for _ in range(N):`
- Generating numeric sequences: `[i**2 for i in range(10)]`
- Iterating by index when you really need the index (but for most cases, `enumerate` — coming next — is better).
