# Exercise 33 — `*args` and `**kwargs`

## Concept

- `*args` collects extra **positional** arguments into a tuple.
- `**kwargs` collects extra **keyword** arguments into a dict.

The names `args` and `kwargs` are convention — the `*` and `**` are what matter.

```python
def summarize(label, *values, **options):
    print(label)
    print("values:", values)        # tuple
    print("options:", options)      # dict

summarize("prices", 10, 20, 30, currency="USD", rounded=True)
# prices
# values: (10, 20, 30)
# options: {'currency': 'USD', 'rounded': True}
```

### Unpacking when calling

The same syntax goes the other way — to **spread** an iterable/dict into a call:

```python
nums = [1, 2, 3]
print(*nums)            # equivalent to print(1, 2, 3)

opts = {"sep": " | ", "end": "!\n"}
print("a", "b", "c", **opts)
```

When to use these:
- Writing thin wrappers that forward unknown args (`def my_log(*args, **kw): logging.info(*args, **kw)`)
- Functions that genuinely accept variadic input (`min(1, 2, 3, 4)` style)
- **Not** as a substitute for clear, named parameters when the call signature is fixed.
