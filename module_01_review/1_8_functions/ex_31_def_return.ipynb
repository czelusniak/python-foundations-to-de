# Exercise 31 тАФ `def`, parameters, `return`

## Concept

```python
def add(a: int, b: int) -> int:
    """Return a + b."""
    return a + b
```

- **Type hints** (`a: int`, `-> int`) are optional at runtime but make the function's contract obvious. Use them.
- A function with no explicit `return` returns `None`.
- A function can return multiple values by returning a tuple: `return min_val, max_val`. The caller usually unpacks: `mn, mx = stats(...)`.
- **Early return** is fine and often cleaner than deep `if/else` nesting.

Naming: functions are verbs (`get_user`, `compute_total`, `parse_line`). Pure functions (no side effects, same input тЖТ same output) are easier to test, compose, and reason about. Prefer them when you can.
