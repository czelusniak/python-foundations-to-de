# Exercise 32 ŌĆö Default arguments (and the mutable default trap)

## Concept

Defaults let you make a parameter optional:

```python
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}"

greet("Alice")              # 'Hello, Alice'
greet("Alice", "Hi")        # 'Hi, Alice'
greet("Alice", greeting="Hi")   # keyword argument ŌĆö clearer for boolean/option-style params
```

### ŌÜĀ’ĖÅ The mutable default trap

Default values are evaluated **once**, when the function is defined ŌĆö not each time it's called. So a mutable default (list, dict, set) is **shared across all calls**:

```python
def bad_append(x, lst=[]):
    lst.append(x)
    return lst

bad_append(1)   # [1]
bad_append(2)   # [1, 2]   ŌåÉ surprise! same list as last call
bad_append(3)   # [1, 2, 3]
```

The fix: use `None` as the default and create the mutable inside.

```python
def good_append(x, lst=None):
    if lst is None:
        lst = []
    lst.append(x)
    return lst
```

This is one of the most common Python footguns. Memorize the pattern.
