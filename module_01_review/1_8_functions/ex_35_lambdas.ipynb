# Exercise 35 — Lambdas (only as `key=`)

## Concept

A lambda is an anonymous, single-expression function:

```python
square = lambda x: x * x
square(3)   # 9
```

Lambdas are **valid Python** but limited: one expression, no statements. The community convention — and the rule we follow in this course — is:

> **Use lambdas only as the `key=` argument to `sorted`, `min`, `max`, and similar helpers.**

For anything else, write a proper `def`. Named functions are more debuggable, more testable, and read better.

```python
rows = [("Alice", 88), ("Bob", 63), ("Carol", 76)]

sorted(rows, key=lambda row: row[1])             # sort by score
max(rows, key=lambda row: row[1])                # ('Alice', 88)
min(rows, key=lambda row: len(row[0]))           # shortest name
```

Why `key=`? The function is throw-away, used once, on the line where you sort. Naming it would add noise without value.
