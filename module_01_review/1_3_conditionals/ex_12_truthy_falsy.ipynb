# Exercise 12 — Truthy and Falsy Values

## Concept

In Python, `if` doesn't require a literal `True` or `False`. Any value is evaluated as truthy or falsy:

**Falsy** (evaluates to `False` in an `if`):
- `0`, `0.0`
- `""` (empty string)
- `[]`, `{}`, `()` (empty collections)
- `None`

**Everything else is truthy.**

```python
name = ""
if name:
    print("Has name")
else:
    print("Name is empty")   # this runs
```
