# Exercise 11 — if / elif / else

## Concept

```python
if condition:
    # runs if condition is True
elif other_condition:
    # runs if the first was False and this is True
else:
    # runs if all above were False
```

Python uses **indentation** (4 spaces) to define blocks — no curly braces.  
Only the first matching branch runs; the rest are skipped.
