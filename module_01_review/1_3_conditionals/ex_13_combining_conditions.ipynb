# Exercise 13 — Combining Conditions

## Concept

```python
# and — both must be True
age >= 18 and age <= 65

# or — at least one must be True
status == "active" or status == "trial"

# not — reverses the result
not is_deleted

# in — membership test (works on strings, lists, dicts)
country in ["BR", "PT", "MZ"]
"@" in email
```

Use parentheses when mixing `and`/`or` to make precedence explicit — `and` binds tighter than `or`.
