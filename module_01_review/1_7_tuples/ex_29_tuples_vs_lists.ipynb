# Exercise 29 — Tuples vs lists

## Concept

A **tuple** is an ordered, **immutable** sequence. Created with parentheses (or just commas):

```python
t = (1, 2, 3)
t = 1, 2, 3            # parens optional
empty = ()
singleton = (42,)      # trailing comma required — (42) is just the int 42
```

Indexing, slicing, `len`, `in` — all work the same as on lists. What you **can't** do is mutate: no `.append`, no item assignment.

### When to prefer tuples

- **Heterogeneous, fixed-structure records:** `(name, age, email)` — different types, fixed positions, like a row from a CSV.
- **Dict keys / set elements:** lists are unhashable, tuples are hashable. You can use `(date, region)` as a composite key.
- **Returning multiple values from a function:** `return min_val, max_val`.
- **Communicates intent:** "this won't change" — a small but real readability win.

Lists are for **homogeneous, variable-length** sequences you build up or mutate.

In DE: rows are often tuples, batches/streams of rows are lists.
