# Exercise 30 — Unpacking and `*rest` patterns

## Concept

Unpacking assigns multiple variables from a sequence in one step:

```python
a, b, c = (1, 2, 3)        # a=1, b=2, c=3
x, y = y, x                # swap, no temp variable
```

The number of names must match the length **unless** you use `*` to absorb the rest into a list:

```python
first, *rest = [1, 2, 3, 4]        # first=1,  rest=[2, 3, 4]
*head, last  = [1, 2, 3, 4]        # head=[1, 2, 3], last=4
first, *middle, last = [1, 2, 3, 4]  # first=1, middle=[2, 3], last=4
```

Unpacking shows up everywhere: `for k, v in d.items():`, `for i, x in enumerate(...):`, returning multiple values, destructuring records.

**Ignore with `_`**: when you don't care about a position, use `_`:

```python
user_id, _, email = (101, "Alice", "a@x.com")    # skip the name
```
