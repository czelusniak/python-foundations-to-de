# Exercise 9 — split() and join()

## Concept

These two methods are complementary — `split` breaks a string into a list, `join` assembles a list back into a string.

```python
# split: str -> list
"a,b,c".split(",")      # ['a', 'b', 'c']
"one two three".split() # ['one', 'two', 'three']  (splits on any whitespace)

# join: list -> str  (note: join is called ON the separator)
",".join(["a", "b", "c"])    # 'a,b,c'
" | ".join(["x", "y", "z"]) # 'x | y | z'
```

In DE, you'll use these constantly to parse delimited text.
