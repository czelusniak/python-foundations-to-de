# Exercise 7 — String Indexing and Slicing

## Concept

Strings are indexed starting at 0. Negative indexes count from the end.

```python
s = "data engineer"
s[0]      # 'd'         — first character
s[-1]     # 'r'         — last character
s[0:4]    # 'data'      — characters 0, 1, 2, 3 (stop is excluded)
s[5:]     # 'engineer'  — from index 5 to the end
s[:4]     # 'data'      — from start to index 4 (excluded)
s[::-1]   # reverse the string
```

Slice syntax: `s[start:stop:step]`
