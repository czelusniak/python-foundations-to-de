# Exercise 10 — f-strings and Formatting

## Concept

f-strings evaluate expressions inside `{}` at runtime:

```python
name = "Alice"
n = 1234567.891

f"Hello, {name}!"          # 'Hello, Alice!'
f"{n:.2f}"                 # '1234567.89'  — 2 decimal places
f"{n:,.2f}"                # '1,234,567.89' — thousands separator
f"{42:05d}"                # '00042'        — zero-padded integer
f"{'left':<10}|"           # 'left      |'  — left-aligned, 10 chars wide
f"{'right':>10}|"          # '     right|'  — right-aligned
```
