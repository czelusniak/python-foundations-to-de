# Exercise 28 ‚Äî Sets

## Concept

A `set` is an unordered collection of **unique, hashable** values. O(1) membership tests ‚Äî much faster than `in` on a list.

```python
s = {1, 2, 3}
s.add(4)
s.discard(2)        # like remove, but no KeyError if missing
2 in s              # O(1)

set([1, 2, 2, 3])   # {1, 2, 3}   ‚Äî easy deduplication
```

**Empty set:** `set()`, not `{}` ‚Äî `{}` is an empty **dict**.

Set algebra (operators or methods):

```python
a = {1, 2, 3}
b = {3, 4, 5}

a | b   # {1, 2, 3, 4, 5}   union
a & b   # {3}               intersection
a - b   # {1, 2}            difference (in a, not in b)
a ^ b   # {1, 2, 4, 5}      symmetric difference
```

When to reach for sets in DE: deduplication, "which records in source A are missing from sink B", `value in allowed_values` checks on large allow-lists.
