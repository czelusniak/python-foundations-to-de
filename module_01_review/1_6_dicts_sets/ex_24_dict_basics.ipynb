# Exercise 24 ŌĆö Dictionary basics

## Concept

A dict maps **keys** to **values**. Keys must be hashable (strings, numbers, tuples ŌĆö not lists). Lookup, insert, and delete are O(1) average.

```python
user = {"id": 101, "name": "Alice", "email": "alice@example.com"}

user["name"]            # 'Alice'           ŌĆö KeyError if missing
user["age"] = 30         # add new key
user["name"] = "Alicia"  # update existing key
del user["email"]        # remove key (KeyError if missing)

len(user)                # number of keys
"name" in user           # True  ŌĆö membership tests the keys
```

From this exercise onwards we use **interview-style framing**: a function signature, a docstring, expected output, and edge cases. Run the cell after you fill it in ŌĆö the `assert`s tell you if it's right.
