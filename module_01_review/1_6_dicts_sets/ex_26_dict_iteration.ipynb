# Exercise 26 — Iterating dicts: `.keys()`, `.values()`, `.items()`

## Concept

Three ways to iterate a dict:

```python
user = {"id": 1, "name": "Alice", "email": "a@x.com"}

for k in user:                # iterates KEYS by default
    print(k)

for k in user.keys():         # explicit — same as above
    print(k)

for v in user.values():       # values
    print(v)

for k, v in user.items():     # pairs — most common
    print(k, v)
```

Dict iteration order is **insertion order** (guaranteed since Python 3.7). Don't rely on alphabetical or any other order.

The views (`.keys()`, `.values()`, `.items()`) are dynamic — they reflect changes to the underlying dict. Don't modify a dict while iterating over it; iterate over `list(d.items())` if you must mutate.
