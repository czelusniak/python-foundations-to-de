# Exercise 27 — `defaultdict` and `Counter`

## Concept

Two specialized dicts from the `collections` module that save you constant boilerplate.

### `defaultdict(factory)`

Like a regular dict, but missing keys are auto-created using `factory()` (a zero-argument callable).

```python
from collections import defaultdict

groups = defaultdict(list)         # missing keys get a fresh []
groups["a"].append(1)              # no need to check if "a" exists
groups["a"].append(2)
groups["b"].append(3)
# defaultdict(<class 'list'>, {'a': [1, 2], 'b': [3]})

counts = defaultdict(int)          # missing keys get 0
counts["x"] += 1
```

### `Counter`

A dict subclass tailored for counting. Pass it any iterable.

```python
from collections import Counter

c = Counter(["a", "b", "a", "c", "b", "a"])
# Counter({'a': 3, 'b': 2, 'c': 1})

c.most_common(2)        # [('a', 3), ('b', 2)]
c["z"]                  # 0   — missing keys return 0, no KeyError
```

In DE you'll use these for: grouping records by a key, counting event types, frequency tables for data quality reports.
