# Exercise 25 — `dict.get()` with default

## Concept

Reading with `d[key]` raises `KeyError` if the key is missing. `.get()` returns a default instead — usually `None`, but you can pass anything:

```python
d = {"a": 1, "b": 2}

d.get("a")          # 1
d.get("c")          # None
d.get("c", 0)       # 0    — explicit default
```

Why this matters in DE: when reading semi-structured JSON, fields may be missing in some records. `.get()` lets you write linear, defensive code without `try/except` everywhere.

Related: `dict.setdefault(key, default)` returns the value if present, otherwise inserts `default` and returns it. Useful for building dict-of-lists patterns (but next exercise's `defaultdict` is usually cleaner).
