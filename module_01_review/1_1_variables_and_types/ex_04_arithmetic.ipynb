# Exercise 4 — Arithmetic Operators

## Concept

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `+`  | addition       | `10 + 3`  | `13`  |
| `-`  | subtraction    | `10 - 3`  | `7`   |
| `*`  | multiplication | `10 * 3`  | `30`  |
| `/`  | division       | `10 / 3`  | `3.333...` (always float) |
| `//` | floor division | `10 // 3` | `3`   (truncates decimal) |
| `%`  | modulo         | `10 % 3`  | `1`   (remainder) |
| `**` | power          | `2 ** 8`  | `256` |

**Key distinction:** `/` always returns a `float`. `//` always returns an `int` (when both operands are ints).

## Task 1 — Basic operations

You have a dataset of 1250 records that need to be processed in batches of 100.

Using only arithmetic operators (no `import`, no functions you haven't seen yet), calculate:

- How many **full batches** will there be?
- How many records are left in the **last (incomplete) batch**?
- What is the **total number of batches** needed (including the incomplete one, if any)?

Print each result with a label, e.g. `Full batches: 12`.

```python
total_records = 1250
batch_size = 100

# your code here
full = (total_records // batch_size)
left = total_records % batch_size
total_batched = full + (1 if left != 0 else 0)


print(f"Full batches: {full}")
print(f"Records left: {left}")
print(f"Including the incomplete one: {total_batched}")
```

## Task 2 — Price calculation

A product costs `R$ 149.90`. A discount of `15%` is applied.

Calculate and print:
- The discount amount (rounded to 2 decimal places)
- The final price (rounded to 2 decimal places)

Use the built-in `round(value, ndigits)` for rounding.

```python
price = 149.90
discount_rate = 0.15

# your code here
discount_number = price*discount_rate
discount_number_round = round(discount_number, 2)
new_price = price - discount_number
new_price_round = round(new_price, 2)

print(discount_number_round)
print(new_price_round)
```
