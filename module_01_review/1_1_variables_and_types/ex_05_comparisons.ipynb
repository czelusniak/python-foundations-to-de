# Exercise 5 — Comparison Operators and Boolean Logic

## Concept

### Comparison operators (return `True` or `False`)

| Operator | Meaning | Example |
|----------|---------|----------|
| `==` | equal to | `5 == 5` → `True` |
| `!=` | not equal to | `5 != 3` → `True` |
| `<`  | less than | `5 < 10` → `True` |
| `<=` | less than or equal | `5 <= 5` → `True` |
| `>`  | greater than | `5 > 3` → `True` |
| `>=` | greater than or equal | `5 >= 10` → `False` |
| `in` | membership | `"a" in "apple"` → `True` |

### Boolean operators (combine conditions)

```python
x > 5 and x < 10      # both must be True
x < 5 or x > 10       # at least one must be True
not (x == 5)          # reverses True/False
```

## Task 2 — String membership

You have a list of allowed email domains. Check if an incoming email belongs to one of them.

Given:
- `email = "alice@company.com"`
- `allowed_domains = "company.com, partner.org, internal.net"`

Check if the email's domain is in the allowed list. Extract the domain part (everything after `@`), then check if it's in the allowed string.

**Hint:** use string methods (like `.split()` and `in`) to check membership.

````python
email = "alice@company.com"
allowed_domains = "company.com, partner.org, internal.net"

# your code here
domain = email.split("@")[1]

allowed_domains_list = allowed_domains.split(",")
print(allowed_domains_list)

print(domain in allowed_domains_list)
````
