# Exercise 1 — Variables and print()

## Concept

A variable is just a name pointing at a value. You create one with `name = value` — no type declaration needed. You display values with `print()`.

```python
x = 10
print(x)            # 10
print("x is", x)    # x is 10
```

````python
# TASK 1: Create a variable called `city` and assign your city's name (a string) to it.
#         Then print it.

city = "Curitiba"
print(city)
````

````python
# TASK 2: Create a variable called `year` and assign the integer 2026 to it.
#         Then print a message like:  "Current year: 2026"

year = 2026
print(f"Current year: {year}")
````

````python
# TASK 3: You can reassign a variable to a new value at any time.
#         Create a variable `count` set to 5, print it, reassign it to 10, print it again.

count = 5
print(count)
count = 10
print(count)
````
