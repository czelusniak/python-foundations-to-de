# Exercise 3 — Type Casting

## Concept

You can convert between types explicitly:

```python
int("42")       # 42
float("3.14")   # 3.14
str(100)        # "100"
bool(0)         # False
```

Values that convert to `False` (called **falsy**): `0`, `0.0`, `""`, `[]`, `{}`, `None`.  
Everything else converts to `True`.

| `raw_price` | `"49.99"`  | `float`                              |
| `raw_flag`  | `"0"`      | `bool` (hint: convert to int first)  |

````python
raw_id = "101"
raw_price = "49.99"
raw_flag = "0"



# your code here
id_converted = int(raw_id)
price_converted = float(raw_price)
flag_converted = bool(int(raw_flag))

print(id_converted, type(id_converted))
print(price_converted, type(price_converted))
print(flag_converted, type(flag_converted))
````
