# Exercise 15 — List Methods

## Concept

```python
lst = [3, 1, 2]

lst.append(4)        # [3, 1, 2, 4]       — add one item to the end
lst.extend([5, 6])   # [3, 1, 2, 4, 5, 6] — add multiple items
lst.insert(0, 99)    # [99, 3, 1, 2, ...]  — insert at index
lst.remove(3)        # removes first occurrence of value 3
lst.pop()            # removes and returns the last item
lst.pop(0)           # removes and returns item at index 0
```

**append vs extend:** `append([5, 6])` adds the list as a single element → `[..., [5, 6]]`.  
`extend([5, 6])` unpacks it → `[..., 5, 6]`.
