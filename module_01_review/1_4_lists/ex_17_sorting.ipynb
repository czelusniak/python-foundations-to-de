# Exercise 17 — Sorting Lists

## Concept

Python has two ways to sort:

```python
lst = [3, 1, 4, 1, 5]

sorted(lst)          # returns a NEW sorted list — original unchanged
lst.sort()           # sorts IN PLACE — modifies original, returns None

sorted(lst, reverse=True)   # descending
```

The `key=` parameter lets you sort by a custom criterion:

```python
words = ["banana", "fig", "apple"]
sorted(words, key=len)         # ['fig', 'apple', 'banana'] — sort by length
sorted(words, key=str.lower)   # case-insensitive sort
```
