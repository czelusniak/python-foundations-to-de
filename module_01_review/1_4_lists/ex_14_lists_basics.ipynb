## Concept

A list is an **ordered, mutable** collection. Unlike strings, lists can be changed in place.

```python
records = [10, 20, 30, 40]

records[0]      # 10   — first element
records[-1]     # 40   — last element
records[1]      # 20

records[0] = 99         # modify in place → [99, 20, 30, 40]
len(records)            # 4

# lists can hold mixed types
row = [101, "Alice", 34, True]
```
