# Exercise 18 — List Built-ins

## Concept

Python has several built-in functions that work on lists (and any iterable):

```python
nums = [4, 1, 7, 2, 9]

len(nums)       # 5    — number of elements
sum(nums)       # 23   — total
min(nums)       # 1    — smallest
max(nums)       # 9    — largest

7 in nums       # True  — membership test
8 in nums       # False

nums.count(1)   # 1   — how many times 1 appears
nums.index(7)   # 2   — first index where 7 appears
```
