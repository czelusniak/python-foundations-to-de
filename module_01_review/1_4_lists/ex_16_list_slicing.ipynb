# Exercise 16 — List Slicing

## Concept

List slicing uses the same `[start:stop:step]` syntax as strings:

```python
records = [10, 20, 30, 40, 50, 60]

records[1:4]    # [20, 30, 40]  — index 1, 2, 3 (stop excluded)
records[:3]     # [10, 20, 30]  — first 3 elements
records[3:]     # [40, 50, 60]  — from index 3 to end
records[-2:]    # [50, 60]      — last 2 elements
records[::2]    # [10, 30, 50]  — every other element
records[::-1]   # [60, 50, 40, 30, 20, 10]  — reversed
```

Slicing a list returns a **new list** — the original is unchanged.
