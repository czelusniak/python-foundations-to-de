# Challenge — Module 1.1: Ship This Order?

**Format:** one problem, one final answer. No mini-tasks. To get the expected output you'll need to combine **every** concept from this folder: variables, the four primitive types, casting, arithmetic operators (`+ - * / // % **`), comparison operators, and boolean operators (`and`, `or`, `not`).

**Scenario.** A warehouse system gives you the raw data for one order (everything arrives as strings, like a CSV row). Your job is to decide whether the order can be shipped today and report the decision plus the numbers that backed it.

## The problem

### Inputs (all strings)

```python
raw_order_total          = "247.85"   # euros
raw_item_count           = "13"       # items in the order
raw_box_capacity         = "4"        # items per shipping box
raw_in_stock             = "1"        # "1" = True, "0" = False
raw_is_priority_customer = "0"        # "1" = True, "0" = False
raw_warehouse_pct_used   = "0.87"     # fraction of warehouse capacity used (0.0–1.0)
raw_truck_box_slots      = "4"        # how many boxes today's truck can carry
```

### Shipping rule

The order ships **only if all four** of these hold:

1. The item is in stock.
2. The order total is at least €100 **OR** the customer is a priority customer.
3. The warehouse is less than 95% full.
4. The total number of boxes needed (full boxes + one extra box if there are leftover items) fits within the truck's slots.

### Expected final output

With the inputs above, a single `print(...)` that produces a line like:

```
will_ship=True | total=247.85 boxes=4 (3 full + 1 leftover) warehouse=0.87
```

(Exact spacing isn't important — but the booleans, numbers, and structure must match.)

### Restrictions

Use **only** what 1.1 covered. No `if` statements, no loops, no functions, no string methods, no f-strings. Just variables, casting, arithmetic, comparisons, and boolean operators.

### Hints (without giving the answer)

- `bool("0")` is `True` — any non-empty string is truthy. To turn `"0"`/`"1"` into a real bool, route through `int(...)` first.
- `//` and `%` together tell you how many full boxes you need and whether there's a leftover.
- "One extra box if there's a leftover" — a comparison returns a bool, and `int(True)` is `1`. That's enough to add the extra box without an `if`.
- The final `will_ship` is just the `and` of four sub-conditions. Build each one as its own variable first; that makes the final expression readable.

````python
raw_order_total          = "247.85"
raw_item_count           = "13"
raw_box_capacity         = "4"
raw_in_stock             = "1"
raw_is_priority_customer = "0"
raw_warehouse_pct_used   = "0.87"
raw_truck_box_slots      = "4"

# your code here — end with a single print(...) of the result line

order_total_float = float(raw_order_total)
item_count_int = int(raw_item_count)
box_capacity_int = int(raw_box_capacity)
in_stock_bool = bool(int(raw_in_stock))
is_priority_customer_bool = bool(int(raw_is_priority_customer))
warehouse_pct_used_float = float(raw_warehouse_pct_used)
truck_box_slots_int = int(raw_truck_box_slots)      

print(in_stock_bool, type(in_stock_bool))

full_boxes = item_count_int // box_capacity_int
leftovers = item_count_int % box_capacity_int
needs_extra_box = (leftovers > 0)
total_boxes=full_boxes+needs_extra_box
print(total_boxes, leftovers)
print("total of boxes:" + str(total_boxes))
print(is_priority_customer_bool, type(is_priority_customer_bool))

print("---")
condition_1 = (in_stock_bool)
condition_2 = (order_total_float>=100 or is_priority_customer_bool)
print(f"condition2: {condition_2}")
condition_3 = (warehouse_pct_used_float < 0.95)
condition_4 = (total_boxes <= truck_box_slots_int)
print(f"condition4: {condition_4}")

will_ship = (condition_1 and condition_2 and condition_3 and condition_4)

print("Will ship = " + str(will_ship) + " | total=" + str(order_total_float) + "  boxes=" + str(total_boxes) + " (" + str(full_boxes) + " full + " + str(leftovers) + " leftover) " + "warehouse=" + str(warehouse_pct_used_float))
````
