# Challenge — Module 1.11: Product Catalog

**Format:** interview-style, multi-class. You must use every concept from 1.11: a plain class with `__init__` and methods, `__repr__`/`__str__`, single inheritance, and `@dataclass`. Plus everything from 1.1–1.10.

**Scenario.** You're modeling a small product catalog with three product kinds. The catalog computes totals, finds the cheapest item, and produces a printable receipt.

## The problem

```python
from dataclasses import dataclass, field

@dataclass
class Product:
    """Base product. MUST use @dataclass.
    Fields: name (str), base_price (float).
    Methods:
      - final_price() -> float: default returns base_price
      - __str__: return 'Product(name=..., price=X.XX)' formatted price
    """
    pass


class DiscountedProduct(Product):
    """Subclass. NOT a dataclass — plain class with __init__ that calls super().__init__.
    Extra field: discount_pct (float between 0 and 1).
    final_price() returns base_price * (1 - discount_pct).
    Override __repr__ to include the discount: 'DiscountedProduct(name=..., -20%)'.
    """
    pass


class BundleProduct(Product):
    """Subclass that wraps a list of Products and applies a bundle_discount on top.
    Extra fields: items (list[Product]), bundle_discount (float).
    base_price for the parent should be 0 — the real price comes from items.
    final_price() = sum(item.final_price() for item in items) * (1 - bundle_discount).
    """
    pass


@dataclass
class Catalog:
    """Holds a list of products. MUST use @dataclass with field(default_factory=list).
    Methods:
      - total() -> float: sum of final_price across all products
      - cheapest() -> Product: the product with the lowest final_price
                              (use min(..., key=lambda p: p.final_price()))
      - receipt() -> str: a multi-line string, one product per line, plus a TOTAL line.
                         Use list comprehension + '\\n'.join.
    Edge cases:
      - Empty catalog: total()=0, cheapest() raises ValueError, receipt()='(empty)'
    """
    pass
```

### Test case

```python
bundle = BundleProduct(name='StarterKit',
                       items=[Product('Pen', 2.00), Product('Notebook', 5.00)],
                       bundle_discount=0.10)

catalog = Catalog(products=[
    Product('Mug', 10.00),
    DiscountedProduct('Shirt', 25.00, discount_pct=0.20),
    bundle,
])

print(catalog.receipt())
print('cheapest:', catalog.cheapest())
```

### Expected output (shape)

```
Product(name=Mug, price=10.00)
DiscountedProduct(name=Shirt, -20%)
BundleProduct(name=StarterKit) -> 6.30
TOTAL: 36.30
cheapest: BundleProduct(name=StarterKit) -> 6.30
```

Math check: Shirt = 25 * 0.8 = 20.00. Bundle = (2+5) * 0.9 = 6.30. Total = 10 + 20 + 6.30 = 36.30.

### Restrictions

Only 1.1–1.11. Must use **both** styles: `@dataclass` for `Product` and `Catalog`, plain classes for `DiscountedProduct` and `BundleProduct`.

````python
from dataclasses import dataclass, field

@dataclass
class Product:
    pass

class DiscountedProduct(Product):
    pass

class BundleProduct(Product):
    pass

@dataclass
class Catalog:
    pass


# Test
bundle = BundleProduct(name='StarterKit',
                       items=[Product('Pen', 2.00), Product('Notebook', 5.00)],
                       bundle_discount=0.10)

catalog = Catalog(products=[
    Product('Mug', 10.00),
    DiscountedProduct('Shirt', 25.00, discount_pct=0.20),
    bundle,
])

print(catalog.receipt())
print('cheapest:', catalog.cheapest())
````
