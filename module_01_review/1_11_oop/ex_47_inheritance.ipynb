# Exercise 47 — Single inheritance basics

## Concept

A subclass extends or specializes a parent class. Declared with the parent in parentheses:

```python
class Animal:
    def __init__(self, name):
        self.name = name
    def speak(self):
        return "..."

class Dog(Animal):
    def speak(self):                          # override
        return "woof"

class Cat(Animal):
    def __init__(self, name, indoor=True):
        super().__init__(name)                # call parent __init__
        self.indoor = indoor
```

### When to inherit (and when not to)

Inheritance is a strong claim: "a `Dog` **is-a** `Animal`". Reach for it when:
- You genuinely have a hierarchy of types that share behavior
- You want to override one method while reusing the rest

Don't inherit just to share a few helper functions — that's **composition**'s job (have an attribute that's the helper). Deep inheritance hierarchies become hard to navigate fast. One level of inheritance is usually enough.

### `super()`

Inside a method, `super()` refers to the parent. Most common use: in `__init__` to call the parent's `__init__` before adding subclass-specific state.

### `isinstance()`

`isinstance(obj, ParentClass)` is `True` for instances of the class or any subclass. Prefer `isinstance` over `type(obj) == ParentClass` — the latter rejects subclasses, which is almost never what you want.
