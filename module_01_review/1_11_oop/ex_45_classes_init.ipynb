## Concept

A class is a blueprint for objects. `__init__` runs once when an instance is created, and sets attributes on `self`:

```python
class User:
    def __init__(self, user_id: int, name: str):
        self.id = user_id
        self.name = name

u = User(1, "Alice")
u.id        # 1
u.name      # 'Alice'
u.name = "Alicia"   # attributes are mutable by default
```

Key points:
- `self` is the first parameter on every method â€” it's the instance the method is called on. You don't pass it; Python does.
- Attributes get created the moment you assign to `self.x`. No "declare-then-assign" needed.
- Calling `User(1, "Alice")` is shorthand for `User.__init__(new_instance, 1, "Alice")`.

### When to use a class vs a dict

- **Dict**: ad-hoc records, JSON-like data, dynamic keys.
- **Class**: fixed set of attributes, methods that operate on them, types you'll pass around.

For pure data with no behavior, `@dataclass` (ex_48) is the modern shortcut. Plain classes are still worth knowing â€” they're the foundation.
