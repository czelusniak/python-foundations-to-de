# Exercise 46 — Methods, `self`, `__repr__` and `__str__`

## Concept

A **method** is just a function defined inside a class — its first parameter is `self`, the instance.

```python
class Counter:
    def __init__(self):
        self.value = 0

    def inc(self, by: int = 1) -> None:
        self.value += by

    def reset(self) -> None:
        self.value = 0
```

### `__repr__` and `__str__`

Both define string forms of an object — they look similar but serve different audiences.

- **`__repr__`** — for developers. Should be unambiguous and ideally look like Python code that would recreate the object. Used by the REPL, debuggers, and the default `print` when `__str__` isn't defined.
- **`__str__`** — for users. Human-friendly. Used by `print(obj)` and `str(obj)`.

Rule: **always define `__repr__`**. Define `__str__` only if you also want a separate user-facing form.

```python
class User:
    def __init__(self, user_id, name):
        self.id = user_id
        self.name = name
    def __repr__(self):
        return f"User(id={self.id!r}, name={self.name!r})"
    def __str__(self):
        return f"{self.name} (#{self.id})"

u = User(1, "Alice")
repr(u)        # "User(id=1, name='Alice')"
str(u)         # 'Alice (#1)'
```

The `!r` in f-strings calls `repr()` on the value — handy inside `__repr__` to get quotes on strings.

- `repr(Money(10.5, "USD"))` → `"Money(amount=10.5, currency='USD')"`
- `str(Money(10.5, "USD"))` → `"USD 10.50"`   (currency code, space, amount with 2 decimals)


Also add a method `add(other: Money) -> Money` that returns a **new** `Money` with the same currency and the summed amount. If currencies differ, raise `ValueError("currency mismatch")`.

````python
class Money:
    pass  # TODO

m = Money(10.5, "USD")
assert repr(m) == "Money(amount=10.5, currency='USD')"
assert str(m) == "USD 10.50"

total = Money(10.5, "USD").add(Money(2.25, "USD"))
assert isinstance(total, Money)
assert total.amount == 12.75
assert total.currency == "USD"

try:
    Money(10, "USD").add(Money(5, "BRL"))
except ValueError as e:
    assert str(e) == "currency mismatch"
else:
    raise AssertionError("expected ValueError")
print("ok")
````
