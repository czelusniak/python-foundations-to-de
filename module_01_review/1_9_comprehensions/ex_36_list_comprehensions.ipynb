# Exercise 36 — List comprehensions

## Concept

A list comprehension builds a list in one expression. Generic shape:

```python
[expression for item in iterable if condition]
```

Three real examples:

```python
squares    = [x * x for x in range(5)]                  # [0, 1, 4, 9, 16]
active     = [u for u in users if u["active"]]          # filter
upper_emails = [u["email"].lower() for u in users]      # transform a field
```

Mental model: read left-to-right as "give me `expr` for each `item`, when `condition`". Equivalent loop:

```python
out = []
for item in iterable:
    if condition:
        out.append(expression)
```

When to prefer comprehensions: short, single-purpose transforms or filters. When to **avoid** them: nested logic, multiple conditions, side effects, or anything that's hard to read on one line. Prefer a regular loop in those cases — clarity beats cleverness.
