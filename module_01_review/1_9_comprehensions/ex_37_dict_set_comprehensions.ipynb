# Exercise 37 — Dict and set comprehensions

## Concept

Same syntax as list comprehensions — the brackets and the expression shape change.

```python
# dict comprehension: key: value pairs
{k: v for k, v in items}
{name: len(name) for name in ["alice", "bob", "carol"]}
# {'alice': 5, 'bob': 3, 'carol': 5}

# set comprehension: a single expression, deduplicated
{x % 3 for x in range(10)}     # {0, 1, 2}
```

Add a filter with `if`:

```python
{u["id"]: u["email"] for u in users if u["active"]}
```

Dict comprehensions are the cleanest way to build lookup tables. Set comprehensions are useful when you want unique results from a transformation.

Note: `{1, 2, 3}` is a **set**, but `{}` is an empty **dict**. For an empty set, use `set()`.
