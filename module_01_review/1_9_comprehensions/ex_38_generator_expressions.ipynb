## Concept

A **generator expression** looks like a list comprehension but uses **parentheses**. It produces values **lazily**, one at a time, instead of building the whole list in memory.

```python
squares_list = [x * x for x in range(10_000_000)]   # builds a 10M-element list
squares_gen  = (x * x for x in range(10_000_000))   # yields one at a time, ~0 memory
```

When iterated, both produce the same values. The difference is **memory** and **eagerness**:

- List comp: builds everything up front. Indexable, reusable.
- Generator: lazy, one-shot (you can only iterate once), `O(1)` memory.

When the next function consumes one value at a time, **drop the brackets**:

```python
total = sum(x * x for x in range(10_000_000))    # parens optional when it's the only argument
any(line.startswith("ERROR") for line in log_lines)
max((row["price"] for row in rows), default=0)
```

In DE this matters when streaming files, API pages, or DB cursors that don't fit in memory.

### One-shot trap

A generator is exhausted after one iteration:

```python
g = (x for x in [1, 2, 3])
list(g)   # [1, 2, 3]
list(g)   # []   â€” already consumed
```

If you need to iterate twice, either store as a list, or rebuild the generator.
